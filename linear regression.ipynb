```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/home/simriti/Desktop/r1.csv')

df=df.dropna()
a=list(df.columns)
X = df[[a[1],a[2],a[3], a[4]]]
Y = df[a[5]]
```

```python
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression


coeff_df = pd.DataFrame(linear_regressor.coef_, X.columns, columns=['Coefficient'])  
print(coeff_df)
```

```python
Y_pred = linear_regressor.predict(X)  # make predictions

print(Y_pred)
```

```python

df4 = pd.DataFrame({'Actual': Y, 'Predicted': Y_pred})
df5 = df.head(5)
print(df5)
```

```python
df5.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
plt.plot(Y,label='Actual')
plt.plot(Y_pred,label='Predicted')
plt.legend()
plt.xlabel('Test Size')
plt.ylabel('Groundwater Level')
plt.title('Groundwater Prediction with Multiple Linear Regression')
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y, Y_pred)))




```
