```python
import pandas as pd
import numpy as np


df=pd.read_csv('/home/simriti/Desktop/r1.csv')

df=df.dropna()
a=list(df.columns)
X = df[[a[1],a[2],a[3], a[4]]]
Y = df[a[5]]




from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

for i in range(0,10):
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.20,random_state = i)
    regressor = RandomForestRegressor(n_estimators=200, max_depth=None, max_features=4, min_samples_leaf=1, min_samples_split=2, bootstrap=False)
    regressor.fit(X_train, y_train)

    predicted1 = regressor.predict(X_test)
    print("Prediction Result: ",predicted1)
    print("i = ",i)
    regr = regressor.score(X_test,y_test)

    print('R-squared test score:', regr) # R-Squared test score

import matplotlib.pyplot as plt
plt.plot(y_test,label='Actual')
plt.plot(predicted1,label='Predicted')
plt.legend()
plt.xlabel('Test Size')
plt.ylabel('Groundwater Level')
plt.title('Groundwater Prediction with Random Forest Regression')
plt.show()
```
